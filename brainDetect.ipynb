```python
%matplotlib inline
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import glob
import os
import random

# Modeli yükle (best.pt yolunu kendine göre düzelt)
model = YOLO(r"C:/Users/EXCALIBUR/Desktop/pasive_project/BrainTumorsDetection/runs/detect/yolov8_tumor_detection_not2/weights/best.pt")

# Test klasöründen rastgele 10 görüntü seç
image_dir = r"C:\Users\EXCALIBUR\Desktop\pasive_project\BrainTumorsDetection\brain_tumor_dataset\images\test"
all_images = glob.glob(os.path.join(image_dir, "*.jpg"))
selected_images = random.sample(all_images, 10)  # Rastgele 10 tane

# 2 satır x 5 sütunluk grid
fig, axes = plt.subplots(2, 5, figsize=(20, 8))

for idx, img_path in enumerate(selected_images):
    results = model.predict(img_path, conf=0.3, verbose=False)
    im = results[0].plot()

    row, col = divmod(idx, 5)
    ax = axes[row, col]
    ax.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    ax.axis("off")
 

plt.tight_layout()
plt.show()
```
